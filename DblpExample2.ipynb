```python
# Sample data
plot_keywords = dataset["movie"].xKeys
simplified_keywords = [keyword.replace('plot_keyword_', '').replace('_', ' ') for keyword in plot_keywords]
X = dataset["movie"].x

# Number of clusters
num_clusters = 20
vectorizer = TfidfVectorizer(stop_words='english')
vectorized_lables = vectorizer.fit_transform(simplified_keywords)

# Perform KMeans clustering
kmeans = KMeans(n_clusters=num_clusters)
kmeans.fit(vectorized_lables)
clusters = kmeans.labels_

# Create a DataFrame with keywords and their assigned clusters
clustered_keywords = pd.DataFrame({'keyword': plot_keywords, 'cluster': clusters})
print(clustered_keywords)

# Create a new DataFrame for the clustered dataset
cluster_columns = [f'cluster_{i+1}' for i in range(num_clusters)]
new_dataset = pd.DataFrame(False, index=range(X.shape[0]), columns=cluster_columns)

# Populate the new DataFrame
for i in range(X.shape[0]):
    for cluster in range(num_clusters):
        # Get keywords in the current cluster
        cluster_keywords = clustered_keywords[clustered_keywords['cluster'] == cluster]['keyword']
        # Set the cluster column to True if any of the keywords in the cluster are True
        if X[i, cluster_keywords.index].any():
            new_dataset.at[i, f'cluster_{cluster+1}'] = True

# Display the new dataset
print(new_dataset)
```
